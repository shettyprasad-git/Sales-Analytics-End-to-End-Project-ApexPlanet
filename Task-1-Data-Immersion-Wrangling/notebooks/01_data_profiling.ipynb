---

### Duplicate Analysis

#### Full Row Duplicates
- No completely duplicated rows were found in the dataset.

#### Order_ID Validation
A detailed investigation was performed on duplicate Order_ID values.

##### Findings
- The dataset contains **992 unique Order_IDs** out of **1000 records**.
- A total of **8 duplicate Order_ID occurrences** were identified.
- All duplicate records share the Order_ID **ORD100050**.
- These records belong to different customers, order dates, products, quantities, and sales amounts.

##### Business Concern
Normally, a single Order_ID should uniquely identify a transaction. The repeated use of the same Order_ID across multiple customers indicates a possible data-entry or system-generated inconsistency.
